## 2. Soluci√≥n via $A=QR$.

Supongamos $A$ de rango m√°ximo para simplificar. Utilizando el m√©todo de Gram-Schmidt, (o via reflexiones de Householder), podemos escribir $A$ como
$$
A=QR,
$$
con $Q$ de Stiefel $R$ triangular superior (e invertible), y en este caso $R$ es invertible.

Luego utilizando las ecuaciones normales tenemos que
$$
A^TA\hat x=A^T b,
$$
por lo que
$$
(QR)^TQR \hat x= R^TQ^Tb,
$$
es decir
$$
R^TQ^TQR \hat x= R^TQ^Tb,
$$
por lo que obtenemos  $R^TR \hat x= R^TQ^Tb,$
 y por lo tanto $$
\hat x= R^{-1}Q^Tb.
 $$



## 4. Optimizaci√≥n





## M√≠nimos Cuadrados como Problema de Optimizaci√≥n

Queremos encontrar $\hat{x}$ que minimice la funci√≥n de p√©rdida cuadr√°tica:

$$
f(x) = \|Ax - b\|^2 = (Ax - b)^T (Ax - b)
$$

### Derivada (gradiente) de $f(x)$

Expandimos:

$$
f(x) = x^T A^T A x - 2 b^T A x + b^T b
$$

Tomamos el gradiente respecto a $x$:

$$
\nabla f(x) = 2 A^T A x - 2 A^T b
$$

Igualando a cero:

$$
\nabla f(x) = 0 \quad \Rightarrow \quad A^T A x = A^T b
$$

üìå ¬°Esta es la ecuaci√≥n normal de m√≠nimos cuadrados!  
La soluci√≥n $\hat{x}$ es el **punto cr√≠tico** (m√≠nimo global, ya que $f$ es convexa).

# M√≠nimos cuadrados y Aprendizaje Estad√≠stico


Supongamos que partimos de un modelo general:
$$
Y = f(X) + \varepsilon,
$$
donde $f$ es cierta funci√≥n en un espacio de funciones generado por ciertas funciones elementales $\varphi_0, \varphi_1, \dots, \varphi_d$.
donde:
$$
f(x) = \sum_{j=0}^{d} \beta_j \varphi_j(x)
$$

Nuestro objetivo es **aprender** $f$ a partir de los datos.

> üî∏ Eventualmente, podr√≠amos tomar un caso a√∫n m√°s general donde f$ est√° en un espacio de dimensi√≥n **infinita**.

---

## Caso del modelo lineal simple

Suponagmos que nuestras funciones elementales son las funciones de $\mathbb R$ en $\mathbb R$ dadas por
$$
\varphi_0(x) = 1, \quad \varphi_1(x) = x
$$
Entonces tenemos:
$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$
Luego a partir de datos podemos estudiar el problema de aprender $\beta_0$ y $\beta_1$. Una posible soluci√≥n es minimizar el error cuadr√°tico medio, y eso nos lleva a un problema de m√≠nimos cuadrados como la clase anterior.


---
## Caso modelo polinomial

An√°logamente a lo anterior, podemos tomar como base de funciones elementales a las funciones monomiales:
$$
\varphi_0(x) = 1, \quad \varphi_1(x) = x,\quad, \varphi_d(x)=x^d.
$$


## Caso general: sistema lineal con funciones base
Ahora podemos considerar un caso un poco m√°s general donde las funciones $\varphi_0,\ldots,\varphi_d$ son arbitrarias (posiblemente ortogonales en alg√∫n producto interno).

Supongamos que tenemos ahora nuestros datos y podemos escribir nuestro sistema
$$
\begin{aligned}
Y_1 &= \beta_0 \varphi_0(x_1) + \dots + \beta_d \varphi_d(x_1) +\varepsilon_1\\\\
Y_2 &= \beta_0 \varphi_0(x_2) + \dots + \beta_d \varphi_d(x_2) \\\\
&\vdots \\\\
Y_m &= \beta_0 \varphi_0(x_n) + \dots + \beta_d \varphi_d(x_m)+\varepsilon_m
\end{aligned}
$$

Luego usando notaci√≥n matricial:
tenemos
$$
Y \in \mathbb{R}^m, \quad \beta \in \mathbb{R}^{d+1}
$$
$$
X =
\begin{pmatrix}
\varphi_0(x_1) & \varphi_1(x_1) & \dots & \varphi_d(x_1) \\\\
\varphi_0(x_2) & \varphi_1(x_2) & \dots & \varphi_d(x_2) \\\\
\vdots & \vdots & & \vdots \\\\
\varphi_0(x_n) & \varphi_1(x_n) & \dots & \varphi_d(x_n)
\end{pmatrix}
\in \mathbb{R}^{m \times (d+1)}
$$

En un modelo de m√≠nimos cuadrados nuestro inter√©s es en estimar $\beta$ de manera tal que minimice el error cuadr√°tico medio, i.e.
resolver el problema de optimizaci√≥n:
$$
\min_{\beta} \|Y - X\beta\|_2
$$


---

## En el modelo lineal cl√°sico:

$$
X =
\begin{pmatrix}
1 & X_1 \\\\
1 & X_2 \\\\
\vdots & \vdots \\\\
1 & X_n
\end{pmatrix}, \quad \varphi_0(x) = 1, \quad \varphi_1(x) = x
$$

---

## Modelo de regresi√≥n polinomial
$$
X =
\begin{pmatrix}
1 & X_{1} & X_{1}^2 & \dots & X_{1}^d \\\\
1 & X_{2} & X_{2}^2 & \dots & X_{2}^d \\\\
\vdots & \vdots & \vdots & & \vdots \\\\
1 & X_{m} & X_{m}^2 & \dots & X_{m}^d \\\\
\end{pmatrix}
$$

---


## Regresi√≥n lineal m√∫ltiple

Las variables de entrada son multivariadas:

$$
X_1 = (X_{11}, X_{12}, \dots, X_{1p}) \\\\
\vdots \\\\
X_n = (X_{n1}, X_{n2}, \dots, X_{np})
$$

Entonces:

$$
X =
\begin{pmatrix}
1 & X_{11} & X_{12} & \dots & X_{1p} \\\\
1 & X_{21} & X_{22} & \dots & X_{2p} \\\\
\vdots & \vdots & \vdots & & \vdots \\\\
1 & X_{n1} & X_{n2} & \dots & X_{np}
\end{pmatrix}
$$

Y se tiene:

$$
Y \approx X \beta \quad \text{(hiperplano en } \mathbb{R}^p \text{)}
$$




```julia
#=
Este notebook muestra c√≥mo ajustar un polinomio a un conjunto de datos mediante m√≠nimos cuadrados.
=#

# Instalar paquetes necesarios
import Pkg
Pkg.add("Plots")
Pkg.add("LinearAlgebra")
Pkg.add("Statistics")

# Cargar paquetes
using Plots
using LinearAlgebra
using Statistics

# Datos a ajustar
x = [0.0, 1.0, 2.0, 3.0, 4.0]
y = [1.0, 2.0, 0.5, 4.0, 3.5]

# Grado del polinomio que queremos ajustar (por ejemplo, cuadr√°tico)
n = 2

# Construir la matriz de Vandermonde
function vandermonde(x, n)
    return [x_i^j for x_i in x, j in 0:n]
end

A = vandermonde(x, n)

# Resolver el sistema sobredeterminado A * c ‚âà y con m√≠nimos cuadrados
c = A \ y

println("Coeficientes del polinomio ajustado: ", c)

# Definir el polinomio ajustado
f(x) = sum(c[j+1] * x^j for j in 0:n)

# Graficar los datos y el polinomio ajustado
xs = range(minimum(x), stop=maximum(x), length=100)
ys = f.(xs)

scatter(x, y, label="Datos", legend=:topleft)
plot!(xs, ys, label="Polinomio ajustado", lw=2)
```

# √Ålgebra Lineal Randomizada: Proyecto!





Este enfoque utiliza vectores aleatorios para hacer m√°s eficientes los c√°lculos en √°lgebra lineal num√©rica, como productos de matrices, descomposici√≥n SVD y QR, especialmente en matrices grandes.

### Motivaci√≥n: Productos Aleatorios $Ax$
Los vectores $Ax$ con $x$ aleatorios son muestras de la subespacio columna de $A$.

### Random Matrix Multiplication

El producto $AB$ requiere $mnp$ multiplicaciones. Si solo muestreamos $s$ columnas $a_k$ de $A$ y filas $b_k^T$ de $B$, podemos aproximar $AB$ con combinaciones de productos externos (suma de matrices de rango 1):
$$
AB \approx \sum_{k=1}^s a_k b_k^T
$$

Para hacerlo m√°s riguroso:

- Sea $S$ una matriz de aleatoria (a elegir).
- Se definen:
  $$
  C = AS, \quad R = S^T B, \quad CR = ASS^T B \approx AB
  $$

La idea es que $\mathbb{E} (SS^T)$ es $\textrm{Id}$, lo cual justifica esta aproximaci√≥n.

La gran idea es que podemos manipular la matriz $S$ para que realmente se tenga $\mathbb{E}(ASS^TB)=AB$ y adem√°s esto ocurra minimanzo la varianza!

Realicemos una breve introducci√≥n y esto puede terminar en otro posible proyecto. Para m√°s referencias, adem√°s de Strang (Secci√≥n II.4) es [Lecture Notes on Randomized Linear Algebra](hhttps://arxiv.org/abs/1608.04481) de Mahoney.

---

## Dos versiones disintas:

### 1. Columnas y filas al azar

- La matriz $S$ tiene una √∫nica entrada no nula por columna, $s_k$.
- $AS$ selecciona columnas de $A$, y $S^T B$ selecciona (las correspondientes) filas de $B$.
- El producto $AS S^T B$ genera una suma de productos fila-columna ponderados:

$$
AB \approx CR = \sum s_k^2 a_k b_k^T
$$

- Cada t√©rmino es de rango 1.
- La selecci√≥n es aleatoria, pero los pesos $s_k$ se eligen apropiadamente.

2. Sampling by random projections

- $S$ tiene varias entradas no nulas por columna.
- As√≠, $AS$ proyecta las columnas de $A$ a menor dimensi√≥n.
- Entonces $AS S^T B$ aproxima de forma m√°s uniforme a $AB$.
